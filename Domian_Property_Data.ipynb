```python
import bs4
import requests
import csv
import fake_useragent
from fake_useragent import UserAgent
import pandas as pd
```

```python
url='https://www.domain.com.au/sold-listings/vic/'
```

```python
page='?page='

for n in range(1, 100):
    if n==1:
        base_url=url
        l=[]
        ua = UserAgent()
        headers = {'User-Agent': ua.random}
        res=requests.get(base_url,headers=headers,timeout=(3.05, 27))
        soup=bs4.BeautifulSoup(res.text,'lxml')
        productDivs = soup.findAll('div', attrs={'class' : "css-qrqvvg"})
        for div in productDivs:
            l.append(div.text)

    
    else:
        base_url=url +page+str(n)
        ua = UserAgent()
        headers = {'User-Agent': ua.random}
        res=requests.get(base_url,headers=headers,timeout=(3.05, 27))
        soup=bs4.BeautifulSoup(res.text,'lxml')
        productDivs = soup.findAll('div', attrs={'class' : "css-qrqvvg"})
        for div in productDivs:
            l.append(div.text)
        
        

```
