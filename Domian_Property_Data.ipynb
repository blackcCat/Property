```python
def extra_info(url, page, h,m,l):
    for n in range(1,50):
        if n==1:
            base_url=url
            l=[]
            sold_date=[]
            ua = UserAgent()
            headers = {'User-Agent': ua.random}
            res=requests.get(base_url,headers=headers,timeout=(3.05, 40))
            soup=bs4.BeautifulSoup(res.text,'lxml')
            productDivs = soup.findAll('div', attrs={'class' : "css-qrqvvg"})
            productDivs1 = soup.findAll('div', attrs={'class' : "css-tmtv67"})
            for div in productDivs:
                l.append(div.text)
            for div1 in productDivs1:
                sold_date.append(div1.text)
        
        else:
            base_url=url +page+str(n)
            ua = UserAgent()
            headers = {'User-Agent': ua.random}
            res=requests.get(base_url,headers=headers,timeout=(3.05, 40))
            soup=bs4.BeautifulSoup(res.text,'lxml')
            productDivs = soup.findAll('div', attrs={'class' : "css-qrqvvg"})
            productDivs1 = soup.findAll('div', attrs={'class' : "css-tmtv67"})
            for div in productDivs:
                l.append(div.text)
            for div1 in productDivs1:
                sold_date.append(div1.text)
                
    Final=list(zip(sold_date,l))

    return Final
```
