# 3. Clean up links

```python
import json
```

```python
retained_links = []

for files in ['rawdata18_04_2025_list1', 'rawdata18_04_2025_list2', 'rawdata18_04_2025_list3', 'rawdata18_04_2025_list4', 'rawdata18_04_2025_list5']:
    with open('../'+files+'.txt', 'r') as f:
        lines = f.readlines()

        for line in lines:
            link_list = eval(line)

            for link in link_list:
                if 'page' in link or 'www.domain.com' not in link or not (link[-1].isdigit()) or 'displaymap' in link:
                    # print(link)
                    pass
                else:
                    retained_links.append(link)
```
